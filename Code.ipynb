```python
# Part 1
import pandas as pd
url = "https://raw.githubusercontent.com/profAltaf/busbis100/main/cellphone_reviews2023.csv"
df = pd.read_csv(url)
df.head()
df.tail()
df.describe()
df.info()
```

```python
# Part 2
unique_brands = df['brand'].unique()
unique_brands
```

```python
# Part 3: Data Cleaning
```

```python
# Part 3 [Converting Data Types To Numeric]
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['totalReviews'] = pd.to_numeric(df['totalReviews'], errors='coerce')
df.info()
df.describe()
```

```python
# Part 3.1 [Remove NA Rows in 'brand']
df.dropna(subset=['brand'],inplace=True)
df.info()
```

```python
# Part 3.2 [Remove Rows Where 'price' = 0]
df_price0 = df.query('price == 0')
df.drop(index = df_price0.index, inplace=True)
df.dropna(subset=['price'],inplace=True)
df.info()
```

```python
# Part 3.3 [Fix 'price' Column]
df['price']
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df_price = df.query('price <= 10')
df.drop(index = df_price.index, inplace=True)
df.info()
```

```python
# Part 3.4 [Fix 'totalReviews' Columns]
df['totalReviews']
df['totalReviews'] = pd.to_numeric(df['totalReviews'], errors='coerce')
df.dropna(subset=['totalReviews'],inplace=True)
df.info()
```

```python
# Part 3.5 [Remove Duplicate asin's]
df.drop_duplicates(subset=['asin'], keep='first',inplace=True)
df.info()
```

```python
# Part 3.6 [Fix 'brand' names]
df['brand'] = df['brand'].str.capitalize()
df['brand'] = df['brand'].replace('samsung','Samsung')
df['brand'] = df['brand'].replace('Moto rola','Motorola')
df['brand'] = df['brand'].replace('Oneplus','OnePlus')
df['brand'] = df['brand'].replace('Asus','ASUS')
df['brand'] = df['brand'].replace('Motolora','Motorola')
unique_brands = df['brand'].unique()
unique_brands
df.head()
```

```python
# Part 3.7 [Delete 'image' and 'originalPrice' Columns]
df.drop(columns =['image','originalPrice'],inplace=True)
df.info()
```

```python
# Part 4: Computed Columns
df['Review_URL'] = ''
Review_URL = "http://www.amazon.com/reviews/"
df['Review_URL'] = Review_URL + df['asin']
df.head()
df['Price (Euros)'] = df['price'] * 0.97
df['Price (Euros)'] = round(df['Price (Euros)'],2)
df.head()
```

```python
# Part 5: Finalizing Cleaned File
df.to_csv('Project_eja64_CleanedData.csv')
unique_brands = df['brand'].unique()
unique_brands
```
